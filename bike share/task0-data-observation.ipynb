```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use("default")
import seaborn as sns
sns.set()
```

```python
bike_df = pd.read_csv("hour.csv")
bike_df.head(3)
```

`cnt` is the target column and others are feature columns. From the dataset, `instant` and `dteday` columns can be ignored

### Feature: `season`

- 0: spring
- 1: summer
- 2: fall
- 3: winter

```python
plt.figure(figsize=(10, 2))
sns.countplot(y="season", data=bike_df)
print(bike_df.season.value_counts().sort_index())
```

### Feature: `yr`

```python
plt.figure(figsize=(10, 1))
sns.countplot(y="yr", data=bike_df)
print(bike_df.yr.value_counts().sort_index())
# 0: 2011, 1: 2012
```

### Feature: `weathersit`

- 1: Clear, Few clouds, Partly cloudy, Partly cloudy
- 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
- 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
- 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog

```python
plt.figure(figsize=(10, 2))
sns.countplot(y="weathersit", data=bike_df)
print(bike_df.weathersit.value_counts().sort_index())
```
